```python
import numpy as np
import pandas as pd

```

```python
data=pd.read_csv('train.csv')
tdata=pd.read_csv('test.csv')
tdata
```

```python
text=np.array(data['text'])
sent=np.array(data['airline_sentiment'])
airline=np.array(data['airline'])
test_text=np.array(tdata['text'])
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
import nltk
```

```python
stop=stopwords.words('english')
punc=list(string.punctuation)
no=list(string.digits)
airline=['southwestair','united',
 'usairways',
 'americanair',
 'virginamerica',
 'delta','jetblue']            
stop+=stop+punc+no
stop
```

```python
file=[]
for i,j in zip(text,sent):
    words=word_tokenize(i)
    temp=[]
    for k in words:
        if k.lower() not in stop and k.isalpha():
            temp.append(k)
    #temp=temp[1:]
    file.append((temp,j))
    
```

```python
test_file=[]
for i in test_text:
    words=word_tokenize(i)
    temp=[]
    for k in words:
        if k.lower() not in stop:
            temp.append(k)
    #temp=temp[1:]
    test_file.append((temp))
```

```python
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
lemm=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean(words):
    out=[]
    for i in words:
        pos=pos_tag([i])
        clean=lemm.lemmatize(i,pos=get_simple_pos(pos[0][1]))
        out.append(clean.lower())
    return out
```

```python
file=[(clean(i),j) for i,j in file]
   
        
```

```python
test_file=[(clean(i)) for i in test_file]
```

```python
test_file
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
training_data=[" ".join(i) for i,_ in file]
testing_data=[" ".join(i) for i in test_file]
```

```python
count=CountVectorizer(max_features=3000)
train=count.fit_transform(training_data)
test=count.transform(testing_data)
```

```python
sent.shape
```

```python
clf=SVC(C=800,gamma=0.001)
#clf=KNeighborsClassifier(n_neighbors=7)
#clf=RandomForestClassifier(n_estimators=25)
clf.fit(train,sent)
```

```python
clf.score(train,sent)
```

```python
y_pred=clf.predict(test)
```
